```python
# Python program to identify
#color in images
  
# Importing the libraries OpenCV and numpy
import cv2
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
import pathlib
%matplotlib inline
```

```python
baseDir = 'rotations-png'
rotDir = os.path.join(baseDir, 'train', '*')
# for x in glob.glob(rotDir):
#     print(x)
imgs = glob.glob(rotDir)
n = np.random.randint(len(imgs))
testImgPath = imgs[n]
plt.title(testImgPath.split('\\')[-1])
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
image = mpimg.imread(testImgPath)
plt.imshow(image)
plt.show()
```

```python
######## K-means clustering ########
Z = image.reshape((-1,3))

# convert to np.float32
Z = np.float32(Z)

# define criteria, number of clusters(K) and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 10
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)

# Now convert back into uint8, and make original image
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((image.shape))

# cv2.imshow('res2',res2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(res2)
```
